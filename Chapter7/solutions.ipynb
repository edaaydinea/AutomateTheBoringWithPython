1. **What is the function that creates Regex objects?**
   - The function that creates Regex objects is `re.compile()`.

2. **Why are raw strings often used when creating Regex objects?**
   - Raw strings (denoted by `r''`) are used to avoid having to escape backslashes, which are common in regex patterns. This makes the pattern easier to read and reduces the chance of errors.

3. **What does the search() method return?**
   - The `search()` method returns a Match object if the pattern is found in the string; otherwise, it returns `None`.

4. **How do you get the actual strings that match the pattern from a Match object?**
   - You can use the `group()` method on the Match object. `group(0)` returns the entire match, while `group(1)`, `group(2)`, etc., return the specific groups defined by parentheses in the regex pattern.

5. **In the regex created from r'(\d\d\d)-(\d\d\d-\d\d\d\d)', what does group 0 cover? Group 1? Group 2?**
   - Group 0 covers the entire match (e.g., `123-456-7890`), group 1 covers the first capturing group (e.g., `123`), and group 2 covers the second capturing group (e.g., `456-7890`).

6. **Parentheses and periods have specific meanings in regular expression syntax. How would you specify that you want a regex to match actual parentheses and period characters?**
   - You would escape them using a backslash: `\(`, `\)`, and `\.`. For example, the regex to match a literal set of parentheses would be `\(\)`.

7. **The findall() method returns a list of strings or a list of tuples of strings. What makes it return one or the other?**
   - `findall()` returns a list of strings if there are no capturing groups in the regex. If there are capturing groups, it returns a list of tuples, where each tuple contains the matched groups.

8. **What does the | character signify in regular expressions?**
   - The `|` character signifies logical OR, allowing for alternative matches. For example, `cat|dog` matches either "cat" or "dog".

9. **What two things does the ? character signify in regular expressions?**
   - The `?` character signifies that the preceding element is optional (i.e., it can occur zero or one time) and can also indicate non-greedy matching when used in patterns like `.*?`.

10. **What is the difference between the + and * characters in regular expressions?**
    - The `+` character signifies that the preceding element must occur one or more times, while the `*` character signifies that it can occur zero or more times.

11. **What is the difference between {3} and {3,5} in regular expressions?**
    - `{3}` specifies that the preceding element must occur exactly three times, while `{3,5}` specifies that it can occur between three and five times (inclusive).

12. **What do the \d, \w, and \s shorthand character classes signify in regular expressions?**
    - `\d` matches any digit (equivalent to `[0-9]`), `\w` matches any alphanumeric character (equivalent to `[a-zA-Z0-9_]`), and `\s` matches any whitespace character (spaces, tabs, newlines).

13. **What do the \D, \W, and \S shorthand character classes signify in regular expressions?**
    - `\D` matches any non-digit character, `\W` matches any non-alphanumeric character, and `\S` matches any non-whitespace character.

14. **What is the difference between .* and .*??**
    - `.*` matches as many characters as possible (greedy matching), while `.*?` matches as few characters as possible (non-greedy matching).

15. **What is the character class syntax to match all numbers and lowercase letters?**
    - The character class syntax to match all numbers and lowercase letters is `[0-9a-z]`.

16. **How do you make a regular expression case-insensitive?**
    - You can make a regex case-insensitive by using the `re.IGNORECASE` flag as the second argument to `re.compile()` or by adding `(?i)` at the start of the pattern.

17. **What does the . character normally match? What does it match if re.DOTALL is passed as the second argument to re.compile()?**
    - The `.` character normally matches any character except a newline. If `re.DOTALL` is passed, it matches any character including newline characters.

18. **If numRegex = re.compile(r'\d+'), what will numRegex.sub('X', '12 drummers, 11 pipers, five rings, 3 hens') return?**
    - It will return `'X drummers, X pipers, five rings, X hens'`.

19. **What does passing re.VERBOSE as the second argument to re.compile() allow you to do?**
    - It allows you to write regex patterns more clearly by ignoring whitespace and allowing comments within the pattern, which helps in documenting complex patterns.

20. **How would you write a regex that matches a number with commas for every three digits?**
    - You can use the following regex: `r'^\d{1,3}(,\d{3})*$'`. This matches a number that starts with one to three digits and may be followed by groups of exactly three digits preceded by commas.

21. **How would you write a regex that matches the full name of someone whose last name is Watanabe?**
    - The regex could be: `r'^[A-Z][a-zA-Z]* Watanabe$'`. This matches a capitalized first name followed by "Watanabe".

22. **How would you write a regex that matches a sentence where the first word is either Alice, Bob, or Carol; the second word is either eats, pets, or throws; the third word is apples, cats, or baseballs; and the sentence ends with a period?**
    - The regex could be: `r'^(Alice|Bob|Carol) (eats|pets|throws) (apples|cats|baseballs)\.$'`. Add the `re.IGNORECASE` flag for case insensitivity.
