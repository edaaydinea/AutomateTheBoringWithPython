1. **Differences between the `webbrowser`, `requests`, `bs4`, and `selenium` modules:**
   - `webbrowser`: Opens a browser and allows you to programmatically control it, e.g., opening a URL.
   - `requests`: A module for sending HTTP requests and handling responses (e.g., downloading web content).
   - `bs4` (BeautifulSoup): A library for parsing HTML and XML documents, useful for web scraping.
   - `selenium`: Automates browser interaction by controlling a browser instance through a script (e.g., clicking buttons, filling forms).

2. **What type of object is returned by `requests.get()`?**  
   - A `Response` object.
   - To access the content as a string: `response.text`.

3. **What requests method checks that the download worked?**  
   - The `response.raise_for_status()` method will raise an exception if the download was unsuccessful.

4. **How can you get the HTTP status code of a `requests` response?**  
   - By accessing the `.status_code` attribute of the `Response` object (e.g., `response.status_code`).

5. **How do you save a `requests` response to a file?**  
   ```python
   with open('file_name', 'wb') as file:
       file.write(response.content)
   ```

6. **What is the keyboard shortcut for opening a browser’s developer tools?**  
   - `F12` or `Ctrl+Shift+I` (Windows/Linux) / `Cmd+Opt+I` (Mac).

7. **How can you view (in the developer tools) the HTML of a specific element on a web page?**  
   - Right-click the element and choose "Inspect" or use the "Elements" tab to browse the DOM tree.

8. **What is the CSS selector string that would find the element with an `id` attribute of `main`?**  
   - `'#main'`.

9. **What is the CSS selector string that would find the elements with a CSS class of `highlight`?**  
   - `'.highlight'`.

10. **What is the CSS selector string that would find all the `<div>` elements inside another `<div>` element?**  
   - `'div div'`.

11. **What is the CSS selector string that would find the `<button>` element with a `value` attribute set to `favorite`?**  
   - `'button[value="favorite"]'`.

12. **How could you get the string `'Hello, world!'` from the BeautifulSoup Tag object `<div>Hello, world!</div>` stored in the variable `spam`?**  
   - `spam.get_text()` or `spam.text`.

13. **How would you store all the attributes of a BeautifulSoup Tag object in a variable named `linkElem`?**  
   - `linkElem.attrs`.

14. **How do you properly import the selenium module?**  
   - `from selenium import webdriver`.

15. **What’s the difference between the `find_element_*` and `find_elements_*` methods?**  
   - `find_element_*`: Returns the first matching element.
   - `find_elements_*`: Returns a list of all matching elements.

16. **What methods do Selenium’s WebElement objects have for simulating mouse clicks and keyboard keys?**  
   - `.click()` for simulating mouse clicks.
   - `.send_keys()` for simulating keyboard input.

17. **An easier way to submit a form with selenium instead of calling `send_keys(Keys.ENTER)` on the Submit button’s WebElement object:**  
   - Use `.submit()` on the form’s WebElement.

18. **How can you simulate clicking a browser’s Forward, Back, and Refresh buttons with selenium?**  
   - `.forward()` to simulate the Forward button.
   - `.back()` to simulate the Back button.
   - `.refresh()` to simulate the Refresh button.
